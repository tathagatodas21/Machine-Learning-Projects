```python
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',encoding='latin-1')
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
ratings.head(2)
```

```python
len(ratings['user_id'].value_counts())
```

```python
ratings['user_id'].value_counts()
```

```python
count=ratings['user_id'].value_counts()
len(count[count>200])
```

```python
#ratings[ratings['user_id'].isin(count[count>200].index)]
count[count>200].index
```

```python
ratings[ratings['user_id'].isin(count[count>200].index)]
```

```python
ratings=ratings[ratings['user_id'].isin(count[count>200].index)]
```

```python
len(ratings)
```

```python
ratings_with_bookname=ratings.merge(book,on='ISBN')
len(ratings_with_bookname)
```

```python
ratings_with_bookname
```

```python
ratings_with_bookname.groupby('Book Name')['rating'].agg('count').reset_index().rename(columns={'rating':'num_rating'})
```

```python
book_num_rating=ratings_with_bookname.groupby('Book Name')['rating'].agg('count').reset_index().rename(columns={'rating':'num_rating'})
```

```python
ratings_with_bookname.merge(book_num_rating, on='Book Name')
```

```python
salman=ratings_with_bookname.merge(book_num_rating, on='Book Name')
```

```python
 srk=salman[salman['num_rating']>=50]
```

```python
srk.drop_duplicates(['user_id','Book Name'],inplace=True)
```

```python
len(srk)
```

```python
amir=srk.pivot_table(index='Book Name',columns='user_id',values='rating').fillna(0)
```

```python
amir.shape
```

```python
from scipy.sparse import csr_matrix
```

```python
ranbir=csr_matrix(amir)
```

```python
type(ranbir)
```

```python
from sklearn.neighbors import NearestNeighbors

nn=NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
nn.fit(ranbir)
```

```python
nn.kneighbors(amir.iloc[0,:].values.reshape(1,888),n_neighbors=6)
```
