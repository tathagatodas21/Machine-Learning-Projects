```python
import numpy as np
import pandas as pd
```

```python
book=pd.read_csv('BX-Books.csv', sep=';', error_bad_lines=False, encoding='latin-1')
```

```python
book=book[['ISBN','Book-Title','Book-Author']]
book.rename(columns={'Book-Title':'Book Name', 'Book-Author':'Author'}, inplace=True)
```

```python
book.head(2)
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=';',encoding='latin-1')
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'}, inplace=True)
ratings.head(2)
```

```python
count=ratings['user_id'].value_counts()
ratings=ratings[ratings['user_id'].isin(count[count>200].index)]
len(ratings)
```

```python
ratings_with_bookname=ratings.merge(book, on='ISBN')
ratings_with_bookname
```

```python
book_num_ratings=ratings_with_bookname.groupby('Book Name')['rating'].agg('count').reset_index().rename(columns={'rating':'num_rating'})
```

```python
salman=ratings_with_bookname.merge(book_num_ratings, on='Book Name')
```

```python
srk=salman[salman['num_rating']>=50]
```

```python
srk.drop_duplicates(['user_id','Book Name'], inplace=True)
```

```python
len(srk)
```

```python
aamir=srk.pivot_table(index='Book Name', columns='user_id', values='rating').fillna(0)
```

```python
aamir.shape
```

```python
aamir.head()
```

```python
from scipy.sparse import csr_matrix
ranbir=csr_matrix(aamir)
```

```python
type(ranbir)
```

```python
from sklearn.neighbors import NearestNeighbors

nn=NearestNeighbors(metric='cosine', algorithm='brute')
```

```python
nn.fit(ranbir)
```

```python
distances,books=nn.kneighbors(aamir.iloc[0,:].values.reshape(1,888), n_neighbors=6)
```

```python
books=books.ravel()
books
```

```python
for i in books:
    print(aamir.index[i])
```
