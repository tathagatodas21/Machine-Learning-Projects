# K Means Clustering

```python
X=data.iloc[:,3:].values
X.shape
```

```python
X
```

```python
from sklearn.cluster import KMeans

wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i,random_state=2)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
    plt.plot(range(1,11),wcss)
```

```python
km=KMeans(n_clusters=5, random_state=2)
y_means=km.fit_predict(X)
y_means
```
